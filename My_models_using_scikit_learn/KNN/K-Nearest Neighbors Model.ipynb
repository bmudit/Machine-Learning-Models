```python
# importing necessary libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import neighbors,metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

#reading the data from the file 
data = pd.read_csv('car.data')

#setting up the necessary variables
x = data[['buying','maint','safety']].values #features
y = data[['class']] #output

#converting the data to numerical values
le = LabelEncoder()
for i in range(len(x[0])):
    x[:,i] = le.fit_transform(x[:,i])
    
#converting the target to nunmerical values
label_maping = {'unacc':0 , 'acc':1, 'good':1 , 'vgood':3}
y['class'] = y['class'].map(label_maping)
y = np.array(y)

# create model
knn = neighbors.KNeighborsClassifier(n_neighbors=25, weights = 'uniform')

# divide data into train and test sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

# train the model
knn.fit(x_train,y_train)

# calculating prediction and accuracy
prediction = knn.predict(x_test)
accuracy = metrics.accuracy_score(y_test,prediction)
```

```python
for i in range(len(prediction)):
    print(f" prediction = {prediction[i]}, original = {y_test[i]}")
```
